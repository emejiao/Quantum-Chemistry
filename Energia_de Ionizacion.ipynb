```python
import psi4
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
import pandas as pd
%matplotlib inline
```

**<font color=green> Inicializamos las condiciones de maquina.** 

```python
# Inicialización Psi4 =================================================================
psi4.core.clean()
psi4.core.clean_options()
psi4.set_memory('2000 MB')  # Memoria que sera utilizada en el cálculo
psi4.set_num_threads(4)    # Cantidad de hilos a utilzar, tener en cuenta que un núcleo soporta 24 hilos. 
                          
psi4.core.set_output_file('./Data/N2_Ionizacion.dat', False) 
# Este comando nos permite guardar el archivo de salida en alguna parte del computador.
```

```python
IPythonConsole.molSize = 50,30
hf = Chem.MolFromSmiles('N#N')
hf
```

**<font color=green> Calculamos la energía de la molécula Neutra previa optimización.** 

```python
# Definir la molécula N2
molecule = psi4.geometry("""
0 1
N
N 1 1.097
""")

# Opciones de cálculo
psi4.set_options({'basis': 'cc-pVDZ', 'scf_type': 'pk', 'reference': 'rhf'
})
# Optimización geométrica
psi4.optimize('scf')

# Energía final
energy_n2 = psi4.energy('scf')

print("Energía electrónica de N2 (Hartree):", energy_n2)
```

**<font color=green> Calculamos la energía de la molécula ionizada, tener encuenta que la multiplicida cambia, $2S+1 = 2 y la carga ahora es +1.** 

```python
# Definición de la molécula N2+
n2_cation = psi4.geometry("""
1 2
N
N 1 1.097
""")

# Opciones de cálculo
psi4.set_options({
    'basis': 'cc-pVDZ',
    'reference': 'uhf',   # sistema abierto
    'scf_type': 'pk'
})
# Optimización geométrica
psi4.optimize('scf')
# Cálculo de energía
energy_n2plus = psi4.energy('scf')

print(f"Energía electrónica de N2+: {energy_n2plus:.3f} Hartree")
```

**<font color=green> El cálculo de la energía de ionización es la diferencia entre la energía de la molécula neutra y la molécula ionizada.** 

```python
EI = energy_n2plus - energy_n2
print(f"Energía de ionización: {EI:.3f} Hartree")
print(f"Energía de ionización:  {EI*27.2114:.3f} eV")
```
